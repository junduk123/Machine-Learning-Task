Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
file_path = '/content/drive/MyDrive/Dataset/StudentsPerformance.csv'
data = pd.read_csv(file_path)

# Daftar kolom bertipe kategorikal
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Fungsi untuk membuat pie chart untuk setiap kolom kategorikal
for column in categorical_columns:
    # Menghitung distribusi persentase untuk setiap kategori
    counts = data[column].value_counts(normalize=True) * 100

    # Membuat pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Persentase Distribusi dari {column}')
    plt.axis('equal')  # Membuat pie chart berbentuk lingkaran
    plt.show()
````

Gender:

Grafik ini menunjukkan distribusi persentase dari jenis kelamin siswa dalam dataset, biasanya dalam kategori "male" dan "female".

Race/Ethnicity:

Pie chart ini menunjukkan persentase siswa berdasarkan kelompok ras atau etnis mereka, yang mungkin dikelompokkan sebagai "Group A", "Group B", dan seterusnya.

Parental Level of Education:

Grafik ini menampilkan distribusi pendidikan terakhir yang ditempuh oleh orang tua siswa, seperti "high school", "associate's degree", "bachelor's degree", dll.

Lunch:

Pie chart ini menunjukkan distribusi jenis makan siang yang diterima siswa, mungkin dikategorikan sebagai "standard" dan "free/reduced".

Test Preparation Course:

Grafik ini menunjukkan distribusi siswa yang mengikuti kursus persiapan ujian, dengan kategori seperti "completed" dan "none".

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_score_influence(df):
    # Set a warm color palette
    sns.set_palette("YlOrRd")

    features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
    scores = ['math score', 'reading score', 'writing score']

    for feature in features:
        for score in scores:
            plt.figure(figsize=(10, 6))
            sns.boxplot(data=df, x=feature, y=score)
            plt.title(f'{score.capitalize()} by {feature.capitalize()}')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

plot_score_influence(df)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_score_distributions(df):
    scores = ['math score', 'reading score', 'writing score']

    plt.figure(figsize=(15, 5))
    for i, score in enumerate(scores, 1):
        plt.subplot(1, 3, i)
        sns.histplot(df[score], kde=True, bins=20, color="skyblue", edgecolor="black")
        plt.title(f'Distribution of {score.capitalize()}')
        plt.xlabel(score.capitalize())
        plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

plot_score_distributions(df)
````

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
import pandas as pd

def calculate_pass_fail(df, threshold=75):
    # Define pass/fail based on the threshold
    pass_fail = {
        'math score': df['math score'] >= threshold,
        'reading score': df['reading score'] >= threshold,
        'writing score': df['writing score'] >= threshold
    }

    # Calculate pass/fail counts for each subject
    results = {}
    for subject, status in pass_fail.items():
        results[subject] = {
            'pass': status.sum(),           # Count of True values (pass)
            'fail': (~status).sum()         # Count of False values (fail)
        }

    # Convert to DataFrame for better visualization
    pass_fail_df = pd.DataFrame(results).T
    pass_fail_df.columns = ['Pass', 'Fail']

    return pass_fail_df

# Get pass/fail counts for each subject
pass_fail_df = calculate_pass_fail(df)
pass_fail_df
````

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_influence_on_scores(df):
    # Set the overall style for the plots
    sns.set(style="whitegrid", palette="pastel")

    # Define the features and scores to plot
    features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
    scores = ['math score', 'reading score', 'writing score']

    # Create box plots for each score by each feature
    for feature in features:
        for score in scores:
            plt.figure(figsize=(10, 6))
            sns.boxplot(data=df, x=feature, y=score)
            plt.title(f'{score.capitalize()} by {feature.capitalize()}')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# Call the function to plot
plot_influence_on_scores(df)
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average score across all subjects for each student
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Define the categorical features to analyze
categorical_features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Plot the average score by each categorical feature
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=feature, y='average score')
    plt.title(f'Average Score by {feature.capitalize()}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Display correlation of average score with each category in a descriptive format
category_summary = df.groupby(categorical_features)['average score'].mean().reset_index()
category_summary
````
